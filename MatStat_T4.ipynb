```python
import random

def G(y : float): # F^-1(y) - обратная к функции распределения
    return 10*y + 10

n = 100
Yn = [random.uniform(0, 1) for i in range(n)]
Xn = [G(y) for y in Yn]

print(Xn)
```

```python
import numpy as np

X = np.array(Xn)

b = 0.95
Xmax = max(X)

teta_ = 2/3 * np.mean(X)
a1 = np.mean(X)
a2 = np.mean(X**2)

u1, u2 = -1.96, 1.96

# Точный доверительный интервал
T1 = Xmax / (((1+b)/2)**(1/n) + 1)
T2 = Xmax / (((1-b)/2)**(1/n) + 1)

# Асимптотический доверительный интервал
A1 = teta_ - 2 * (a2 - a1**2)**0.5 * u2 / (3*n**0.5)
A2 = teta_ - 2 * (a2 - a1**2)**0.5 * u1 / (3*n**0.5)

print('Точный доверительный интервал: (', T1, ';', T2, '), l =', T2-T1)
print('Асимптотический доверительный интервал: (', A1, ';', A2, '), l =', A2-A1)
```

```python
# bootstrap: Доверительный интервал (непараметрический)
import random

delta = []
for i in range(1000):
    Xn_ = []
    for j in range(len(Xn)):
        Xn_.append(random.choice(Xn))
    teta_star = 2/3 * np.mean(Xn_)
    delta.append(teta_star - teta_)

delta.sort()

k1 = int((1-b)/2 * 1000)
k2 = int((1+b)/2 * 1000)

delta_k1 = delta[k1 - 1]
delta_k2 = delta[k2 - 1]

B1 = teta_ - delta_k2
B2 = teta_ - delta_k1

print('Бутстраповский доверительный интервал: (', B1, ';', B2, '), l =', B2 - B1)
```
