```python
import random

def G(y : float): # F^-1(y) - обратная к функции распределения
    return (1/(1-y)) ** (1/2)

n = 100
Yn = [random.uniform(0, 1) for i in range(n)]
Xn = [G(y) for y in Yn]

print(Xn)
```

```python
import numpy as np
from math import log

X = np.array(Xn)

b = 0.95

teta_ = 1/np.mean(np.log(X)) + 1

def g(teta):
    return 2 ** (1/(teta-1))

u1, u2 = -1.96, 1.96

# Доверительный интервал для медианы
M1 = g(teta_) - u2 * 2**(1/(teta_-1)) * log(2) / (n**0.5 * (teta_-1))
M2 = g(teta_) - u1 * 2**(1/(teta_-1)) * log(2) / (n**0.5 * (teta_-1))

# Асимптотический доверительный интервал для teta
A1 = teta_ - u2 * (teta_-1) / n**0.5
A2 = teta_ - u1 * (teta_-1) / n**0.5

print('Доверительный интервал для медианы: (', M1, ';', M2, '), l =', M2-M1)
print('Асимптотический доверительный интервал для teta: (', A1, ';', A2, '), l =', A2-A1)
```

```python
# bootstrap: Доверительный интервал для teta (непараметрический)
import random

delta = []
for i in range(1000):
    Xn_ = []
    for j in range(len(Xn)):
        Xn_.append(random.choice(Xn))
    teta_star = 1/np.mean(np.log(Xn_)) + 1
    delta.append(teta_star - teta_)

delta.sort()

k1 = int((1-b)/2 * 1000)
k2 = int((1+b)/2 * 1000)

delta_k1 = delta[k1 - 1]
delta_k2 = delta[k2 - 1]

B1 = teta_ - delta_k2
B2 = teta_ - delta_k1

print('Бутстраповский непараметрический доверительный интервал:\n (', B1, ';', B2, '), l =', B2 - B1)
```

```python
# bootstrap: Доверительный интервал для teta (параметрический)
N = 5000

def G(y, teta): # F^-1(y, teta) - обратная к функции распределения
    return (1/(1-y)) ** (1/(teta-1))

Teta = []
for i in range(N):
    Xn_ = []
    for j in range(len(Xn)):
        y = random.uniform(0, 1)
        Xn_.append(G(y, teta_))
    teta_star = 1/np.mean(np.log(Xn_)) + 1
    Teta.append(teta_star)

Teta.sort()

k1 = int((1-b)/2 * N)
k2 = int((1+b)/2 * N)

B1 = Teta[k1 - 1]
B2 = Teta[k2 - 1]

print('Бутстраповский параметрический доверительный интервал:\n (', B1, ';', B2, '), l =', B2 - B1)
```
